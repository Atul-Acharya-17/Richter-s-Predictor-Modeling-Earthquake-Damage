Import the essential libraries

*numpy* for linear algebra\
*pandas* for data manipulation\
*matplotlib* for visualizing data\

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sb
sb.set() 
```

Adjust the path accordingly

```python
PATH = ''
```

```python
X = pd.read_csv(PATH + 'train_values.csv')
y = pd.read_csv(PATH + 'train_labels.csv')
```

```python
X.head()
```

```python
y.head()
```

```python
print("Data type : ", type(X))
print("Data dims : ", X.shape)
```

```python
print("Data type : ", type(y))
print("Data dims : ", y.shape)
```

```python
dataset = pd.merge(X, y, on='building_id')
```

```python
print("Data type : ", type(dataset))
print("Data dims : ", dataset.shape)
```

##### Check the type of each attribute

```python
dataset.dtypes
```

##### Check for null values

```python
dtypes=pd.DataFrame(dataset.dtypes,columns=["Data Type"])
dtypes["Unique Values"]=dataset.nunique()
dtypes["Null Values"]=dataset.isnull().sum()
dtypes["% of null Values"]=dataset.isnull().sum()/len(dataset)
dtypes
```

### The dataset doesn't have any missing values
---

```python
dataset.describe().T
```

```python
categorical = pd.DataFrame(dataset[['land_surface_condition', 
                                    'foundation_type', 
                                    'roof_type', 
                                    'ground_floor_type', 
                                    'other_floor_type', 
                                    'plan_configuration',
                                    'position', 
                                    'legal_ownership_status',
                                    'count_families']])

binary1 = pd.DataFrame(dataset[['has_superstructure_adobe_mud',
                               'has_superstructure_mud_mortar_stone',
                               'has_superstructure_stone_flag', 
                               'has_superstructure_cement_mortar_stone',
                               'has_superstructure_mud_mortar_brick',
                               'has_superstructure_cement_mortar_brick',
                               'has_superstructure_timber',
                               'has_superstructure_bamboo',
                               'has_superstructure_rc_non_engineered', 
                               'has_superstructure_rc_engineered',
                               'has_superstructure_other']])

binary2 = pd.DataFrame(dataset[["has_secondary_use",
                               'has_secondary_use_agriculture',
                               'has_secondary_use_hotel',
                               'has_secondary_use_rental',
                               'has_secondary_use_institution',
                               'has_secondary_use_school',
                               'has_secondary_use_industry',
                               'has_secondary_use_health_post',
                               'has_secondary_use_gov_office',
                               'has_secondary_use_use_police',
                               'has_secondary_use_other']])

numeric = pd.DataFrame(dataset[["age",
                                "area_percentage",
                                "height_percentage",
                               "geo_level_1_id",
                               "geo_level_2_id",
                               "geo_level_3_id"]])
```

```python
dataset["damage_grade"].describe()
```

## Standard Visualization of the Predictors 

```python
plt.figure(figsize=(28,8))
sb.countplot(x=dataset["area_percentage"],hue=dataset["damage_grade"]).set_title("Area% vs Damage Grade")
plt.ylabel("Number of Buildings")
plt.xlabel("Area Percentage")
```

```python
plt.figure(figsize=(15,8))
sb.countplot(x=dataset["height_percentage"],hue=dataset["damage_grade"]).set_title("Height% vs Damage Grade")
plt.ylabel("Number of Buildings")
plt.xlabel("Height Percentage")
```

```python
for column in numeric:
    plt.figure(figsize=(15,5))
    sb.boxplot(x = numeric[column], y = y["damage_grade"], orient = "h")
    plt.show()
```

### Quite a few outliers exist in the dataset and seem to be a bit skewed

```python
dataset_copy = dataset[dataset["age"]<300]
plt.figure(figsize=(15,5))
sb.boxplot(x = dataset_copy["age"], y = dataset_copy["damage_grade"], orient = "h")
```

### Understadning the distribution of the geo locations

```python
f, axes = plt.subplots(3, 1, figsize=(24, 12))
sb.distplot(dataset['geo_level_1_id'], bins = 21, kde = True, ax = axes[0], color = 'red')
sb.distplot(dataset['geo_level_2_id'], bins = 23, ax = axes[1], color = 'green')
sb.distplot(dataset['geo_level_3_id'], bins = 21, ax = axes[2], color = 'blue')
```

---
### Jointplot to check if theres any strong correlation between the geo locations

```python
sb.jointplot(x=dataset["geo_level_1_id"], y=dataset["geo_level_2_id"], height=8)
sb.jointplot(x=dataset["geo_level_1_id"], y=dataset["geo_level_3_id"], height=8)
sb.jointplot(x=dataset["geo_level_3_id"], y=dataset["geo_level_2_id"], height=8)
```

### No correlations can be seen

### Distribution of categorical variables

```python
font={'weight' : 'bold', 'size': 16}       
q = 1
plt.figure(figsize=(20,20))
for j in categorical:
    plt.subplot(3,3,q)
    ax = sb.countplot(dataset[j].dropna(),palette="hls")
    plt.xticks(**font)
    plt.yticks(**font)
    plt.xlabel(j)
    plt.legend(["Low damage","Average damage","High damage"])
    q+=1
plt.show()
```
